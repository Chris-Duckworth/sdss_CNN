# create_dataset  

Creating input image datasets along with corresponding labels.

--- 

- Input data is found from SDSS photometry of centred on the ra and dec of the target galaxy. Images are taken from sciserver (see in `sdss_CNN/data/on_sciserver`), so that images are 424 x 424 pixels, and, cover a sky area equal to 8.48R<sub>e</sub> (elliptical petrosian half radii)
, so that each pixel covers 0.02R<sub>e</sub>. 

- These are pre-processed for CNN input here by scaling all flux values by the maximum value in the image, and reshaped (downscaled) to (80, 80, 3).

- Labels are computed from MaNGA kinematics. Galaxy spin is the label of choice, estimated by the Î»<sub>r</sub> parameter. 

```python
import numpy as np
import matplotlib.pyplot as plt
import progressbar
import time
import sdss_prep
from PIL import Image
import os
import pandas as pd
```

### selecting arbitrary galaxy to return image for

```python
im_path = '../../data/on_sciserver/images/7443-12701cutout.jpeg'
plt.imshow(sdss_prep.process_image(im_path))
```

### loading all usable galaxies with lambda_R meaurements.

```python
tab = pd.read_csv('/Users/chrisduckworth/astronomy/projects/spin_bias/catalogues/manga.Pipe3D_v2_5_3.csv')
tab.plateifu.values, tab.lambda_re.values
```

### iterating over all columns in pandas.df, appending to x and y only if image file exists.

```python
x = []
y = []

for pifu in tab.plateifu.values:
    try :
        filepath = '../../data/on_sciserver/images/' + pifu + 'cutout.jpeg'
        im = sdss_prep.process_image(filepath, shape=(80, 80))
        x.append(im)
        y.append(pifu)
        
    except :
        print(pifu + ' missing.')
```

### saving output as np.array objects.

```python
outpath = '../../data/cleaned_datasets/'

np.save(outpath + 'x', x)
np.save(outpath + 'y', y)
```
